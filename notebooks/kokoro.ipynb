# Kokoro

```python
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
from datetime import datetime


def kokoro_generate_audio(text, CONFIG: Configuration):
    pipeline = KPipeline(lang_code=CONFIG.kokoro_language)

    generator = pipeline(
        text, 
        voice=CONFIG.kokoro_voice,
        speed=CONFIG.kokoro_play_speed,
    )
    for i, (gs, ps, audio) in enumerate(generator):
        print(i, gs, ps)
        display(Audio(data=audio, rate=CONFIG.kokoro_rate, autoplay=i==0))

        save_path = os.path.join(
            CONFIG.OUTPUT_PATH, 
            f'kokoro-{CONFIG.kokoro_voice}-{CONFIG.kokoro_language}-{datetime.today().strftime("%Y-%m-%d")}.wav'
        )

        sf.write(save_path, audio, CONFIG.kokoro_rate)
```

```python
text = "Ruiiiz, donde estas carlitos, jummmm que ganas tengo de comerte a besos mi amor."

kokoro_generate_audio(text, CONFIG)
```
